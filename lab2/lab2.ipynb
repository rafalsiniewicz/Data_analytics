```python
import scipy.stats as stats
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#1
#create samples
np.random.seed(1000)
fig, axs = plt.subplots(3, 3, figsize=(15,15))
mu = (100, 50, 10)
sigma = (15, 50, 100)
n = 0
samples = []
for i in range(len(mu)):
    for j in range(len(sigma)):
        samples.append(mu[i] + sigma[j] * np.random.randn(10000))
        axs[i][j].plot(samples[n])
        axs[i][j].set_title('mu= {}, sigma= {}'.format(mu[i], sigma[j]))
        n += 1
```

```python
#histogram
fig, axs = plt.subplots(3, 3, figsize=(15,15))
n = 0
for i in range(len(mu)):
    for j in range(len(sigma)):
        axs[i][j].hist(samples[n], bins=sigma[j])
        axs[i][j].set_title('mu= {}, sigma= {}, bins={}'.format(mu[i], sigma[j], sigma[j]))
        n += 1
```

```python
#cumulative histogram
fig, axs = plt.subplots(3, 3, figsize=(15,15))
n = 0
for i in range(len(mu)):
    for j in range(len(sigma)):
        axs[i][j].hist(samples[n], bins=sigma[j], cumulative=True)
        axs[i][j].set_title('mu= {}, sigma= {}, bins={}'.format(mu[i], sigma[j], sigma[j]))
        n += 1
```

```python
#2
from scipy.special import logit, expit
inverse_logit = expit(samples)
```

```python
fig, axs = plt.subplots(3, 3, figsize=(15,15))
n = 0
for i in range(len(mu)):
    for j in range(len(sigma)):
        axs[i][j].plot(inverse_logit[n])
        axs[i][j].set_title('mu= {}, sigma= {}'.format(mu[i], sigma[j]))
        n += 1
```

```python
#histogram
fig, axs = plt.subplots(3, 3, figsize=(15,15))
n = 0
for i in range(len(mu)):
    for j in range(len(sigma)):
        axs[i][j].hist(inverse_logit[n], bins=sigma[j])
        axs[i][j].set_title('mu= {}, sigma= {}, bins={}'.format(mu[i], sigma[j], sigma[j]))
        n += 1
```

```python
#cumulative histogram
fig, axs = plt.subplots(3, 3, figsize=(15,15))
n = 0
for i in range(len(mu)):
    for j in range(len(sigma)):
        axs[i][j].hist(inverse_logit[n], bins=sigma[j], cumulative=True)
        axs[i][j].set_title('mu= {}, sigma= {}, bins={}'.format(mu[i], sigma[j], sigma[j]))
        n += 1
```
