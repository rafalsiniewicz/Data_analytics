```python
import scipy.stats as stats
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

<center>
    <h1>Normal distribution</h1>
    <br>
    Creating samples basing on mu and sigma parameters
</center>

```python
#1
#create samples
np.random.seed(1000)
fig, axs = plt.subplots(3, 3, figsize=(15,15))
mu = (100, 50, 10)
sigma = (15, 50, 100)
n = 0
samples = []
for i in range(len(mu)):
    for j in range(len(sigma)):
        samples.append(mu[i] + sigma[j] * np.random.randn(10000))
        axs[i][j].plot(samples[n])
        axs[i][j].set_title('mu= {}, sigma= {}'.format(mu[i], sigma[j]))
        n += 1
```

<center>
    <h1>Histograms for normal distribution</h1>
    <br>
    Plotting histograms for different mu and sigma and bins. As we can see depending on mu and sigma value normal distribution shifts left or right to 0. For big mu and small sigma it shifts to right and for small mu and big sigma the most values are oscillating near 0.
</center>

```python
#histogram
fig, axs = plt.subplots(3, 3, figsize=(15,15))
n = 0
for i in range(len(mu)):
    for j in range(len(sigma)):
        axs[i][j].hist(samples[n], bins=sigma[j])
        axs[i][j].set_title('mu= {}, sigma= {}, bins={}'.format(mu[i], sigma[j], sigma[j]))
        n += 1
```

<center>
    <h1>Cumulative histograms for normal distribution</h1>
    <br>
    Creating cumulative histograms for different mu, sigma and bins. Histograms differs from range- the bigger sigma, the bigger the range is. 
</center>

```python
#cumulative histogram
fig, axs = plt.subplots(3, 3, figsize=(15,15))
n = 0
for i in range(len(mu)):
    for j in range(len(sigma)):
        axs[i][j].hist(samples[n], bins=sigma[j], cumulative=True)
        axs[i][j].set_title('mu= {}, sigma= {}, bins={}'.format(mu[i], sigma[j], sigma[j]))
        n += 1
```

<center>
    <h1>Normal distribution_inverse_logit</h1>
    <br>
    Creating samples basing on mu and sigma parameters
</center>

```python
#2
from scipy.special import logit, expit
inverse_logit = expit(samples)
```

```python
fig, axs = plt.subplots(3, 3, figsize=(15,15))
n = 0
for i in range(len(mu)):
    for j in range(len(sigma)):
        axs[i][j].plot(inverse_logit[n])
        axs[i][j].set_title('mu= {}, sigma= {}'.format(mu[i], sigma[j]))
        n += 1
```

<center>
    <h1>Plotting histograms for inverse logit distribution</h1>
    <br>
    As we can see below histograms differs from normal distribution. Here most of the values are located on the edges
</center>

```python
#histogram
fig, axs = plt.subplots(3, 3, figsize=(15,15))
n = 0
for i in range(len(mu)):
    for j in range(len(sigma)):
        axs[i][j].hist(inverse_logit[n], bins=sigma[j])
        axs[i][j].set_title('mu= {}, sigma= {}, bins={}'.format(mu[i], sigma[j], sigma[j]))
        n += 1
```

```python
#cumulative histogram
fig, axs = plt.subplots(3, 3, figsize=(15,15))
n = 0
for i in range(len(mu)):
    for j in range(len(sigma)):
        axs[i][j].hist(inverse_logit[n], bins=sigma[j], cumulative=True)
        axs[i][j].set_title('mu= {}, sigma= {}, bins={}'.format(mu[i], sigma[j], sigma[j]))
        n += 1
```

<center>
    <h1>Poisson distribution</h1>
    <br>
    In Poisson distribution mean and variance are equal, i.e. the variance = mean, so we will change only mean as poison
</center>

```python
#3 
from numpy.random import poisson
mean = [5, 10, 15, 20, 50, 100, 1000]
poisson_dist = []
for i in mean:
    poisson_dist.append(poisson(i, 10000))

fig, axs = plt.subplots(len(mean), 1, figsize=(10,30))
n = 0
import matplotlib.pyplot as plt
for i in range(len(mean)):
    axs[i].hist(poisson_dist[n], bins=15)
    axs[i].set_title('mean, variance= {}'.format(mean[i]))
    n += 1
    
#cumulative
fig, axs = plt.subplots(len(mean), 1, figsize=(10,30))
n = 0
import matplotlib.pyplot as plt
for i in range(len(mean)):
    axs[i].hist(poisson_dist[n], bins=15, cumulative=True)
    axs[i].set_title('cumulative: mean, variance= {}'.format(mean[i]))
    n += 1
```

```python
#4 
# beta natural
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
import scipy.stats
alpha = (1, 2, 5, 10, 20, 50, 100, 500, 1000)
beta = (1, 2, 5, 10, 20, 50, 100, 500, 1000)
fig, axs = plt.subplots(len(alpha), len(alpha), figsize=(30,50))

for i, a in enumerate(alpha):
    for j, b in enumerate(beta):
        axs[i][j].hist(np.random.beta(a,b,10000), bins=100)
        mean, var, skew, kurt = scipy.stats.beta.stats(a, b, moments='mvsk')
        x = np.linspace(scipy.stats.beta.ppf(0.01, a, b), scipy.stats.beta.ppf(0.99, a, b), 100)
        axs[i][j].plot(x, scipy.stats.beta.pdf(x, a, b), 'r-', lw=5, alpha=0.6, label='beta pdf')
        axs[i][j].set_title('alpha= {}, beta= {}'.format(a, b))
        
```

```python
# beta location dispersion parametrization
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
import scipy.stats
disp = (0.1, 0.5, 0.9)
loc = (0.1, 0.5, 0.9)
alpha = []
beta = []
for i, d in enumerate(disp):
    for j, l in enumerate(loc):
        alpha.append((1 - d)*l/d)
        beta.append((1 - d)*(1-l)/d)
        
fig, axs = plt.subplots(len(alpha), len(alpha), figsize=(30,50))

for i, a in enumerate(alpha):
    for j, b in enumerate(beta):
        axs[i][j].hist(np.random.beta(a,b,10000), bins=100)
        mean, var, skew, kurt = scipy.stats.beta.stats(a, b, moments='mvsk')
        x = np.linspace(scipy.stats.beta.ppf(0.01, a, b), scipy.stats.beta.ppf(0.99, a, b), 100)
        axs[i][j].plot(x, scipy.stats.beta.pdf(x, a, b), 'r-', lw=5, alpha=0.6, label='beta pdf')
        axs[i][j].set_title('alpha= {}, beta= {}'.format(a, b))
```

```python
# beta mean-population size parametrization
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
sizes=[2, 5, 10, 20, 100, 200, 1000]
alphas=[]
betas=[]
for n in sizes:
    alpha=0.485*n
    beta=n-alpha
    alphas.append(alpha)
    betas.append(beta)

theta1=np.linspace(0.3,0.6,200)
fig, axs = plt.subplots(len(alphas), len(betas), figsize=(30,50))
for i, a in enumerate(alphas):
    for j, b in enumerate(betas):
        s = np.random.beta(438+a,544+b,1000)
        axs[i][j].hist(s, bins=100)
        axs[i][j].plot(theta1,stats.beta.pdf(theta1,438+a,544+b), color='r')
        axs[i][j].axis('tight')
        axs[i][j].set_title('alpha= {}, beta= {}'.format(a, b))
        
fig, axs = plt.subplots(len(alphas), len(betas), figsize=(30,50))
for i, a in enumerate(alphas):
    for j, b in enumerate(betas):
        s = np.random.beta(438+a,544+b,1000)
        axs[i][j].hist(s, bins=100, cumulative=True)
        axs[i][j].plot(theta1,stats.beta.pdf(theta1,438+a,544+b), color='r')
        axs[i][j].axis('tight')
        axs[i][j].set_title('cumulative: alpha= {}, beta= {}'.format(a, b))
```

```python
#5 lognorm
import matplotlib.pyplot as plt
# mean and standard deviation
mu = [0.1, 0.3, 0.5, 1., 2., 5., 10.]
sigma = [0.1, 0.3, 0.5, 1., 2., 5., 10.] 
fig, axs = plt.subplots(len(mu), len(mu), figsize=(30,50))
for i, m in enumerate(mu):
    for j, sig in enumerate(sigma):
        s = np.random.lognormal(m, sig, 1000)
        count, bins, ignored = axs[i][j].hist(s, bins=100)
        x = np.linspace(min(bins), max(bins), 10000)
        pdf = ((np.exp(-(np.log(x) - m)**2 / (2 * sig**2))) / (x * sig * np.sqrt(2 * np.pi)))
        axs[i][j].plot(x, pdf, linewidth=2, color='r')
        axs[i][j].axis('tight')
        axs[i][j].set_title('mu= {}, sigma= {}'.format(m, sig))
```
